```python
def operator_login():
    max_attempts = 3
    correct_user_id = "operator"
    correct_password = "password"
    
    for attempt in range(max_attempts):
        user_id = input("Enter your user ID: ")
        password = input("Enter your password: ")
        
        if user_id == correct_user_id and password == correct_password:
            print("Operator login successful!")
            return True
        else:
            remaining_attempts = max_attempts - (attempt + 1)
            if remaining_attempts > 0:
                print(f"Login failed. {remaining_attempts} attempts remaining.")
    
    print("Login failed. No more attempts remaining.")
    return False

def check_vaccination_status():
    mobile_number = input("Enter your mobile number: ")
    
    # Simulated database of citizens and vaccination status
    citizens = {
        "1234567890": {"vaccinated": False, "first_dose_date": None},
        "9876543210": {"vaccinated": True, "first_dose_date": "2023-05-15", "second_dose_date": "2023-08-15"}
        
    }
    
    citizen = citizens.get(mobile_number)
    
    if not citizen:
        print("Citizen not found.")
    else:
        if not citizen["vaccinated"]:
            print("You haven't taken your first vaccine dose yet. Thank you for visiting us!")
        else:
            first_dose_date = citizen["first_dose_date"]
            second_dose_date = citizen.get("second_dose_date")
            
            if not second_dose_date:
                print("Congratulations! You are fully vaccinated.")
            else:
                from datetime import datetime, timedelta
                current_date = datetime.now().date()
                next_dose_date = datetime.strptime(second_dose_date, "%Y-%m-%d").date()
                days_until_next_dose = (next_dose_date - current_date).days
                
                if days_until_next_dose > 0:
                    print("You are not eligible for your second dose.")
                else:
                    print("You are eligible for your second dose.")
                    print(f"Next dose date: {next_dose_date}")

# Main program
if operator_login():
    check_vaccination_status()
```
